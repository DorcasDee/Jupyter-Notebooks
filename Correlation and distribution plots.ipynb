## Calculate the Body Mass Index (BMI), defined by BMI = P/A 2 (kg/m 2

### **Body Mass Index** is a simple calculation using a person’s height and weight. The formula is $BMI = \frac{kg}{m^2}$ where $kg$ is a person’s weight in kilograms and $m^2$ is their height in metres squared.

```python
heights = fil['Alt']
weights = fil['Pesos']
n = len(fil['Idade'])
BM = []

for i in range (0, n):
    height = heights[i]
    weight = weights[i]
    BMI = weight/height**2
    print('The Body Mass Index', i, 'is', BMI, 'kg/m2')
    BM.append(BMI)
```

```python
fil['BMI'] = BM
fil
```

## For each histogram, plot the Gaussian function on it and classify the result in terms of asymmetry and kurtosis

The **skewness** and **kurtosis** parameters are both measures of the shape of the distribution. 

### **Skewness (coefficient of asymmetry) gives information about the tendency of the deviations from the mean to be larger in one direction than in the other.** 

A positive value for skewness indicates that the data are skewed to the right; a negative value indicates that the data are skewed to the left. 

In other words the skewness coefficient measures the departure from symmetry. The normal distribution has skewness equal to zero.

### **Kurtosis is primarily a measure of the heaviness of the tails of a distribution. The normal distribution has a kurtosis equal to 3 and is recognized as _mesokurtic_**. For this reason, some sources use the following definition of kurtosis (often referred to as "excess kurtosis").

$K = \frac{\mu^4}{\sigma^4}-3$

This definition is used so that the _standard normal distribution_ has a kurtosis of zero.

The kurtosis represents a generalization of the mean and variance measures used for location and dispersion for a Gaussian distribution.

An increased kurtosis (>3) can be visualized as a thin “bell” with a high peak whereas a decreased kurtosis corresponds to a broadening of the peak and “thickening” of the tails. Kurtosis >3 is recognized as **_leptokurtic_** and <3 as **_platykurtic_** (lepto=thin; platy=broad).

```python
# Gaussian function

fig, ax = plt.subplots(figsize =(12, 9))

mmean = fil['Idade'].mean()
Stddev = fil['Idade'].std()
n = len(fil['Idade'])

# Fit a normal distribution to the data:
# mean and standard deviation
mean, std = norm.fit(fil['Idade']) 
  
# Plot the histogram.
plt.hist(fil['Idade'], bins=10, density=True, alpha=0.6, color='b')
  
# Plot the PDF.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 50)
p = norm.pdf(x, mean, std)
  
plt.plot(x, p, 'g', linewidth=2)
plt.title('Age plot with its gaussian function',size=16)
plt.xlabel('Age',size=16)
plt.ylabel('Frequency',size=16)

plt.show()
```

## Perform a linear adjustment on each plot of the previous item (e) and calculate the correlation coefficient between these variables. Discuss which of these graphs have a good correlation between variables

### **Correlation is a statistical metric for measuring to what extent different variables are interdependent.**

In other words, when we look at two variables over time, if one variable changes how does this affect change in the other variable?

This time we'll _visualize two variables using a scatter plot and an added linear line called a regression line, which indicates the relationship between the two._


**One way to measure the strength of the correlation between *continuous numerical variable* is by using a method called *Pearson correlation*.**

#### The Pearson’s correlation coefficient is **calculated as the covariance of the two variables divided by the product of the standard deviation of each data sample**. It is the normalization of the covariance between the two variables to give an interpretable score.

Pearson's correlation coefficient = $\frac{covariance(X, Y)}{stdv(X) * stdv(Y)}$

where:

    Covariance is calculated as the average of the product between the values from each sample, where the values haven been centered (had their mean subtracted).

    The calculation of the sample covariance is as follows:


    $cov(X, Y) = (\sum (x - (\bar X)) * (y - (\bar Y)) ) * 1/(n-1)$
    
    The cov() NumPy function can be used to calculate a covariance matrix between two or more variables.

    "covariance = cov(data1, data2)"

**The use of the mean in the calculation suggests the need for each data sample to have a Gaussian or Gaussian-like distribution.**

Pearson correlation method will give two values: 

1. the correlation coefficient 
2. the P-value.



### Interpretation of these values

#### For the correlation coefficient, 

The coefficient returns a value between -1 and 1 that represents the limits of correlation from a full negative correlation to a full positive correlation. 

1. value close to +1: large positive correlation
2. value close to -1: large negative correlation
3. value close to 0: no correlation between the variables

#### The P-value will tell us how certain we are about the correlation that we calculated. For the P-value, 

1. **$P-value < .001$:** **strong certainty** about the correlation coefficient that we calculated.
2. **$.001 < P-value < .05$:** **moderate certainty**.
3. **$.05 < P-value < .1$:** **weak certainty**.
4. **$P-value > .1$** no certainty of correlation at all. 

#### strong correlation 
- when the correlation coefficient is close to 1 or negative 1
- the P-value < .001. 


```python
# Height versus weight linear regression
from scipy import stats
fig, ax = plt.subplots(figsize =(12, 9))

ax = sns.regplot(x = "Pesos", y = "Alt", data = fil, color='blue', marker='h')

# correlation
pearson_coef, p_value = stats.pearsonr(fil["Pesos"], fil["Alt"])

print('The pearson correlation coefficient is: ',pearson_coef)
print('The P-value is: ',p_value)
```

The straight line through the data points is **steep** which shows that there's a **positive linear relationship** between the two variables (Height and Weight). This shows that Height can **predict** Weight.

The **correlation coefficient** is approximately .7 and this is close to 1

**_There is a strong positive correlation_**

Also, the **P-value** is very small, much smaller than .001

**_So we are certain about the strong positive correlation_**
