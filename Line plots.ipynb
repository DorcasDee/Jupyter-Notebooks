# 3. Simple Chart

```python
#create an empty figure with no axes
fig = plt.figure()

# add axes to the figure
axes = fig.add_axes([0.1, 0.1, 1.5, 1.5]) # left, bottom, width, height - (range from 0 to 1)

#plot on specified axes
axes.plot(x,y, 'm--h', mec = 'b', ms = 19, mfc = 'none') 

# m--h - m = magenta colored line, -- for dashed line, h = hexagon marker, 
# mec = marker edge color: r - red, ms = marker size: 19 - larger dark box, 
# mfc = marker face color: none - removed the dark and changed to transparent

axes.set_xlabel("X-axis", fontsize = 15) # Note the use of set to start methods.
axes.set_ylabel("Y-axis", fontsize = 15)
axes.set_title("The title", fontsize = 15) #The default font is smaller
axes.grid()

plt.savefig("Graphic examples.png", bbox_inches = "tight") # save the file in the folder
plt.show()
```

# 4. Two Graphics + Axis configuration 

```python
# Create an empty figure
fig = plt.figure()

# Add axes to the figure
axes1 = fig.add_axes([0, 0, 1.5, 1.5]) # principal axes
axes2 = fig.add_axes([1.6, 1, 0.4, 0.4]) #Inserted axes

# Working on the major axis
axes1.plot(x,y, "g")
axes1.set_xlabel("X-axis", fontsize = 14)
axes1.set_ylabel("Y-axis", fontsize = 14)
axes1.set_title("Title 1", fontsize = 14)
#axes.grid()

# Working on the minor axis
axes2.plot(x, y, "m")
axes2.set_xlabel("X-axis", fontsize = 8)
axes2.set_ylabel("Y-axis", fontsize = 8)
axes2.set_title("Title 2", fontsize = 8)
axes2.set_ylim([0,4]) #scaling the y axis
axes2.set_xlim([4,10]) #scaling the x-axis

plt.show()
```

# 5. Two Graphics with legend 

```python
# Create a empty figure, criar uma figura vazia
fig = plt.figure(facecolor = "y", edgecolor = "b")

#Add axes to the figure
axes = fig.add_axes([0.1, 0.1, 1.5, 1.5]) # left, bottom, width, height - (range from 0 to 1)

# Plot 1
axes.plot(x,y, "b-.v", mec = "g", ms = 10, mfc = "none", label = "Graph 1")
# Plot 2
axes.plot(x-1, y-1, "r-v", mec = "g", ms = 8, mfc = "g", label = "Graph 2")

axes.set_xlabel("X-axis") #Note the use of set
axes.set_ylabel("Y-axis")
axes.set_title("Title")

axes.legend()

plt.show()
```

# 6. Additional axes 

```python
fig, ax1 = plt.subplots(figsize = (8,5), dpi = 150)

color1 = "red"

ax1.plot(x, x**2, color = color1)
ax1.set_xlabel("X-axis")
ax1.set_ylabel(r"x²", fontsize = 20, color=color1)
ax1.spines["left"].set_color(color1)
ax1.tick_params(axis='y', labelcolor=color1) #color the y-axis labels

ax2 = ax1.twinx() #Creates a second y-axis that shares the same x-axis.
color2  = 'blue'

ax2.set_ylabel(r'x³', fontsize=20, color=color2)
ax2.plot(x, x**3, color=color2)
ax2.tick_params(axis='y', labelcolor=color2) #colr the y axis labels
ax2.spines['right'].set_color(color2)
ax2.spines['left'].set_color('none')

fig.tight_layout()
plt.show()
```

# 7. Subplots and scale 

https://matplotlib.org/stable/api/axes_api.html

```python
#Plot variable
t = np.arange(0.01, 20.0, 0.01)
print(t)

# Figure creation
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize = (10,8))

# log y axis
ax1.semilogy(t, np.exp(-t/5.0))
ax1.set(title = 'semilogy')
ax1.grid(ls='--')

#log x axis
ax2.semilogx(t, np.sin(2 * np.pi *t))
ax2.set(title = 'semilogx')
ax2.grid()

# log x and y axis
ax3.loglog(t, 20 * np.exp(-t/10.0))
ax3.set_yscale('log')
ax3.grid()

#scatter
x = 10.0 **np.linspace(0.0, 2.0, 20)
y = x**2.0

ax4.scatter(x,y, marker='*', c = 'red')
ax4.set_yscale('log')
ax4.grid()

fig.tight_layout()
plt.show()
```

```python
np.random.seed()

#example data
mu = 100 # mean of distribution
sigma = 15 # standard deviation of distribution
x = mu + sigma * np.random.randn(437)

num_bins = 50

fig, ax = plt.subplots(figsize = (10,8))

#plot
n, bins, patches = ax.hist(x, num_bins, density = True)

#add a line of best fit

best_fit_line = ((1 / (np.sqrt(2*np.pi) * sigma)) * np.exp(-0.5 * (1 / sigma*(bins-mu))**2)) #is this a formula?
ax.plot(bins, best_fit_line, '--')

ax.set_xlabel("Smartness")
ax.set_ylabel("Probability density")
ax.set_title(r"IQ Histogram: $\mu = 100, \sigma = 15$")

fig.tight_layout()
plt.show()
```

# 10. Exercise 

https://matplotlib.org/3.1.1/gallery/misc/zorder_demo.html

```python
x = np.linspace(-5,10,100)
x_v1 = 1
x_v2 = 1
y_v1 = 0
y_v2 = -4

x_m = np.mean([x_v1, x_v2])
y_m = np.mean([y_v1, y_v2])

fig, ax = plt.subplots(figsize = (20,10))

#Define the origin of the axes
ax.spines['right'].set_color('none')
ax.spines['top'].set(color = 'none')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

y1 = -3*x**2 + 6*x - 3
y2 = x**2 - 2*x -3
ax.plot(x, y1, lw=2, label=(r'$g(x)=-3x²+6x-3$'), zorder=1) #lw - line width: line thickness
ax.plot(x, y2, lw=2, label=(r'$f(x)=x²-2x - 3$'), zorder=2)

ax.set_xlim([-2, 4])
ax.set_ylim([-5, 5])
#ax.set(xlabel='x', ylabel='y', title='Exercise')
#ax.set_xticks([-2.-1.0,1,2,3,4,5])

ax.plot([x_v1,x_m,x_v2], [y_v1,y_m,y_v2],'m--p', ms=15, zorder=3) #zorder signifies the order with which the lines will be draen over each other
ax.axhline(y=-2, color='g', linestyle='-', label=(r'$reta \quad r$'), zorder=1)

plt.legend()
plt.savefig('fig.png')
plt.show()
```
