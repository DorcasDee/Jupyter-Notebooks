# 1. Central Tendency or Central Value: 

In statistics, a central tensency or measure of central tendency is a central or typical value for a probability distribution.

The most common measures of central tendency are the arithmetic mean, the median and the mode. A central tendency can be calculated for either a finite set of values or for a theorectical distribution such as the normal distribution.

# Measure of Central Tendency
A measure of central tendency is a single value that attempts to describe a set of data by identifying the central position within that set of data. The are also classed as _summary statistics_.

## 1.1 Mean

### It is the most popular and well known measure of central tendency. It can be used with both discrete and continuous data, although it is commonly used with continuous data. The mean is equal to the sum of data divided by number of values in the data set.

### One of the major disadvantage of mean is that it is susceptible to the influence of outliers.

Outlier is an observation point that is distant from other observations. It may be due to variability in the measurement or indicate experimental measurement error.

$mean = \bar{x} = \frac{x_1 + x_2 + x_3 + ... + x_n}{n}$

The **sample** mean: 

$mean = \bar{x} = \frac{\sum x}{n}$

The **population** mean:

$mean = \mu = \frac{\sum x}{n}$

**Example:** _Find the mean of numbers; 1,2,3,4,5,6,7._

**Solution:** _Total = 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28_

_n = 7_

_Average or Mean = $ \frac{28}{7}= 4 $_

## 1.2 Median

### It is the value that lies in the middle of a data that has been arranged in order of magnitude. It divides the frequency distribution exactly into two halves. Hence median is the 50th percentile and also known as 'positional average.

#### Advantages
1. It is easy to compute and comprehend.

2. It is not distorted by outliers/skewed data.

#### Disadvantages
1. It does not take into account the precise value of each observation and hence does not use all information available in the data.

2. Unlike mean, median is not amenable to further mathematical calculation and hence is not used in many statistical tests.

3. If we pool the observations of two groups, median of the pooled group cannot be expressed in terms of the individual medians of the pooled groups.


It can also be derived in several ways



Method 1: **Calculating with python function**

1. Sort the sample. 
The sample can be sorted in ascending or descending order

2. Locate the value in the middle of the data

When locating the median, two things can happen
1. If the length of the data is **odd** (i.e. $\frac{(n+1)^{th}}{2}$  observations), the median is middle value.

**Example:** _FInd the median of 1,2,3,4,5._ 

_The median is 3_

To print the median of an odd set: (i) rearrange the data sample, (ii) find the length of the sample, n = 5, (iii) floor division of n, n//2 = 2, (iv) index = 2, i.e. x[2] gives the median

2. If the length of the data is **even**, the median is the average of the 2 middle values (i.e the $\frac{n^{th}}{2}$ and $(\frac{n}{2} + 1)^{th}$ observation).



**Example:** _Find the median of 1,2,3,4,5,6_

_The middle numbers are 3,4_

The median is $\frac{(3+4)}{2} = 3.5$ 

(ii) length of sample, n= 6, (iii) floor division, n//2 = 3, (iv) index = 2 and 3 gives the middle numbers, i.e $ \frac{x[n-1]+x[n]}{2} $

```python
def amedian(x):
    n = len(x)
    index = n // 2
    #print(n)
    if (n%2) != 0: #data with odd length
        med = sorted(x)[index]
        return med
    else:          #data with even length
        med1 = sorted(x)[index-1]
        med2 = sorted(x)[index]
        med_avg = (med1+med2)/2
        #return med1, med2
        return med_avg
    
    #return sum(sorted(x)[index-1 : index])/2


amedian(x)
```

## 1.3 Mode

### It is the most frequent observation(s) in a data set. Some data sets may not have a mode because each value occur once while some data sets can have more than one mode.

### Mode is rarely used as a summary statistic except to describe a bimodal distribtuion. In a bimodal distribution, the taller peak is the major mode and the shorter one os the minor mode.

#### Advantages
1. It is the only measure of central tendency that can be used for data measured in a nominal scale.[5]

2. It can be calculated easily.

#### Disadvantages
1. It is not used in statistical analysis as it is not algebraically defined and the fluctuation in the frequency of observation is more when the sample size is small.


Commonly used for categorical data sets seuch as _boolean_, _nominal_, _ordinal_

Method1: **Using Python Function**

the collections.counter class in python has the .most_common() method which returns more common elements and their counts.

```python
from collections import Counter

def amode(x):
    a = Counter(x)
    #return a
    return [k for k, v in a.items() if v == a.most_common(1)[0][1]]
            
    

amode(x)
```

### Summary: Position of measures of central tendency

The relativ position of the three measures of central tendency considered above (mean, median, mode) depends on the shape of the distribution. All three measures are identical in a normal distribution.

Since mean is always pulled toward the extreme observations, the mean is shifted towards the tail in a skewed distribution.

Mode, being the most frequently occuring value, lies in the lump of the skewed distribution.

Median lies in the middle of the mean and mode in a skewed distribution.

https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3157145/figure/F1/?report=objectonly

### SELECTING THE APPROPRIATE MEASURE

Mean is generally considered the best measure of central tendency and the most frequently used one. However, there are some situations where the other measures of central tendency are preferred.

Median is preferred to mean when

1. There are few extreme scores in the distribution.

2. Some scores have undetermined values.

3. There is an open ended distribution.


Mode is the preferred measure when data are measured in a nominal scale. 

Geometric mean is the preferred measure of central tendency when data are measured in a logarithmic scale.

# 2. Measures of Dispersion 

## 2.1 Variance

### In a population, variance is the average of the **squared** deviation from the population mean

Formula:
$\sigma² = \frac{\sum f(X-\mu)²}{N}$ where $\sigma²$ is the population variance, $N$ is the number of ekements in the population, $\mu$ is the population mean.

### Sample variance uses a slightly different notation

$s² = \frac{\sum f(x-\bar{x})²}{n-1}$ where $s²$ is the sample variance, $n$ is the number of element in the sample.


## 2.2 Standard Deviation

(Deviation simply means how far from the normal)

### Standard deviation is a measure of how spread out numbers are about the mean value, symbolised by $\sigma$. It is the square root of variance.

### It is useful in comparing sets of data which may have thesame mean but a different range (i.e. highest number-lowest number). 

The standard deviation of a population is

$Standard deviation = \sigma = \sqrt{\frac{\sum f(X-\mu)²}{N}}$ where $\sigma$ is the population standard deviation, $\mu$ is the population mean.

Statisticans often use simple random samples to estimate the standard deviation of a population based on sample data.

$Standard deviation = s = \sqrt{\frac{\sum f(x-\bar{x})²}{n-1}}$

Example: Find the variance and standard deviation of the data below

|x | frequency | x-mean      | (x-mean)²      |  f(x-mean)²     |
|:-:|:--------:|:-----------:|:--------------:|:---------------:|
| 2 |   1      |   -1        |         1      |        1        |
| 3 |   2      |   0         |         0      |        0        |
| 4 |   2      |   1         |         1      |        2        |
|   |  N = 5   |             |                |    $\sum = 3$ |


$mean = \bar{x} = \frac{1(2)+2(3)+2(4)}{5} = \frac{15}{5} = 3$

$variance = \sigma² = \frac{\sum f(x-\bar{x})²}{\sum f} = \frac{3}{5} = 0.6$

$Standard deviation = \sigma = \sqrt{\frac{\sum f(x-\bar{x})²}{\sum f}} = \sqrt{0.6} = 0.774597$

Note: We have **grouped data** and **ungrouped data** and the method of calculation is different. The example above is a grouped data. Grouped data can also be given in groups/class.

Also, there are two formulas for Standard Deviation

- **Population** Standard Deviation: $\sigma = \sqrt{\frac{\sum f(x-\bar{x})²}{\sum f}}$
- **Sample** Standard Deviation: a sample data is a selection from the bigger population. $s = \sqrt{\frac{\sum f(x-\bar{x})²}{\sum f-1}}$

```python
from numpy import sqrt
from numpy import mean

def avariance(x):
    y = abs(x - amean(x))**2
    vari = mean(y)
    return vari

avariance(x)
```

## Absolute and relative dispersion 

They are two different ways to measure the spread of a data set.

### Absolute measure of dispersion includes:
- Range
- Quartile deviation
- Mean deviation
- Standard deviation
- Variance

### Relative measure of dispersion are calculated as ratios and percentages.

They are useful for making comparisons between seperate data sets or different experiments that might use different units.

They are sometimes called _coefficients of dispersion_.

## 2.3 Range

It is the spread of data from the lowest to the highest level of distribution, commonly used measure of variability.

Formula: $Range = H - L$

where H is highest value and L is lowest value

Steps:

- Order all values in the data
- subtract the lowest from the highest

Because only two numbers are used, the range is easily influenced by outliers. It can't describe the shape of the distibution on its own.

To get a clear idea of the data's variability, it is better used in combination with other measures of variability such as interquartile range and standard deviation.

```python
def arange(x):
    return max(x) - min(x)

arange(x)
#print(min(x))
#print(max(x))
```

## 2.4 Quartile Deviation

The Quartile Deviation (QD) is the product of half of the difference between the upper and lower quartiles. Mathematically we can define as:

$Quartile Deviation = \frac{(Q3 – Q1)}{2}$

```python
Q1 = np.quantile(x, 0.25)
print('Quartile 1:', Q1)
Q3 = np.quantile(x, 0.75)
print('Quartile 3:',Q3)

def QD(a,b):
    return(a-b)/2

print('Quartile Deviation:', QD(Q1,Q3))
```

```python
QD(40, 70)
```

## 2.5 Mean Deviation

### It is the mean of the distances of each value from the mean

$Mean Deviation = \frac{\sum |x-\bar{x}|}{N}$

```python
#N = len(x)
def meandev(x):
    y = abs(x - amean(x))
    
    return mean(y)

meandev(x)
```

## 2.6 Coefficient of variation

### It is a measure of relative variability. It is the ratio of the standard deviation to the mean (average). 

Example: the expression “The standard deviation is 15% of the mean” is a CV.

$Coefficient Variation = \left(\frac{\sigma}{\bar{x}} \right) \times 100 $

```python
def CV(x):
    b = (standarddev(x)/amean(x)) * 100
    return b
    
CV(x)
print("The standard deviation is ", CV(x), 'of the mean')
```

## 3.4. Box plot

### A boxplot summarizes the distribution of a numeric variable for one or several groups. It allows to quickly get the median, quartiles and outliers but also hides the dataset individual data points. 



```python
np.random.seed(10)
data1 = np.random.normal(100, 20, 200) #loc(mean) = 100, size = 200, scale(standard deviation, spread or width)=20
data2 = np.random.normal(80, 10, 200)
data3 = np.random.normal(120, 30, 200)
data4 = np.random.normal(50, 15, 200)
data = [data1, data2, data3, data4]

# green_diamond = dict(markerfacecolor='g', marker='D')

fig = plt.figure(figsize =(12, 7))
plt.boxplot(data, notch=True)

plt.show()
```

```python
print(data)
```

### Seaborn is one of the most widely used data visualization libraries in Python, as an extension to _Matplotlib._ Seaborn is definitely the best library to quickly build a boxplot. It offers a dedicated boxplot() function that roughly works as follows:

Turn normal arrays into dataframes.

Each of the DataFrames will have a variable called datum.

The data in the score variables will be the data from the normally distributed Numpy arrays we just created in the last step.

Each DataFrame will also have a variable called class.

The class variable will denote which “class” the data is from: 'Class 1', 'Class 2', or 'Class 2'.

Like data set from a school with different classeswith different values from each classes.

This class variable will serve as a categorical variable that we can use to split out our data.

```python
dataA = pd.DataFrame({'datum':data1, 'class':'class1'})
dataB = pd.DataFrame({'datum':data2, 'class':'class2'})
dataC = pd.DataFrame({'datum':data3, 'class':'class3'})
```

#### Combine data together into dataframe

Next, we will concatenate these three DataFrames together using the pd.concat() function.

```python
classdata = pd.concat([dataA, dataB, dataC])
```

#### Add variable gender using pandas assign method

```python
classdata = classdata.assign(gender = np.where(classdata.datum%3 > 1, "Male","Female"))
```

```python
sns.set()
```

```python
sns.boxplot(data=classdata, x='datum')
```

Mean is roughly around 90. Maximum is about 110, minimum is about 78

First quartile marked out at the left side 

Third quartile marked out at the right side 

This give a sense of how the data is distributed

#### Changing the color of the seaborn plot. Default is blue

```python
sns.boxplot(data=classdata, x='datum', color='magenta')
```

#### Same plot different color

#### Breaking out into the categorical variable _'class'_

```python
sns.boxplot(data=classdata,
           x='datum',
           y='class',
           color='magenta'
           )
```

#### Now there is one box for each class. Now we will use the hue parameter to change the color of the categorical variable _'gender'_

```python
sns.boxplot(data=classdata,
           x='datum',
           y='class',
           hue='gender'
           )
```

#### Data is broken into two categorical variables, male and female.

#### And by putting _'class'_ on the y axis, the data was broken out by _'class'_

#### Let us try and change the orientation by swapping the axis

```python
sns.boxplot(data=classdata,
           x='class',
           y='datum',
           color='magenta'
           )
```
